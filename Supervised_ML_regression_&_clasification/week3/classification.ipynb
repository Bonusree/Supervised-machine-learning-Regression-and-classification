```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
x_train = np.array([0., 1, 2, 3, 4, 5])
y_train = np.array([0,  0, 0, 1, 1, 1])
X_train2 = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train2 = np.array([0, 0, 0, 1, 1, 1])
```

```python
# Define positive and negative classes for single-variable plot
pos = y_train == 1
neg = y_train == 0

# Colors for consistent style
dlc = {"dlblue": "#1f77b4"}

# Define the plot_data function for two-variable data
def plot_data(X, y, ax):
    pos = y == 1
    neg = y == 0
    ax.scatter(X[pos, 0], X[pos, 1], marker='x', c='red', s=80, label="y=1")
    ax.scatter(X[neg, 0], X[neg, 1], marker='o', facecolors='none', edgecolors=dlc["dlblue"], 
               s=100, label="y=0", lw=2)

# Create subplots
fig, ax = plt.subplots(1, 2, figsize=(8, 3))

# Plot 1: Single-variable plot
ax[0].scatter(x_train[pos], y_train[pos], marker='x', s=80, c='red', label="y=1")
ax[0].scatter(x_train[neg], y_train[neg], marker='o', s=100, label="y=0", facecolors='none',
              edgecolors=dlc["dlblue"], lw=3)
ax[0].set_ylim(-0.08, 1.1)
ax[0].set_ylabel('y', fontsize=12)
ax[0].set_xlabel('x', fontsize=12)
ax[0].set_title('One-variable plot')
ax[0].legend()

# Plot 2: Two-variable plot
plot_data(X_train2, y_train2, ax[1])
ax[1].axis([0, 4, 0, 4])  # Set x and y axis limits
ax[1].set_ylabel('$x_1$', fontsize=12)
ax[1].set_xlabel('$x_0$', fontsize=12)
ax[1].set_title('Two-variable plot')
ax[1].legend()

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
