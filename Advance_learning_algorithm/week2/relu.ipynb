```python
def relu_activation(z):
    if z>0:
        return z;
    else:
        return 0;
```

```python
set_w = np.array([[2]])  # Weight: 1x1 matrix
set_b = np.array([-4.5])  # Bias: 1x1 array
X_train = np.array([0., 1, 2, 3, 4, 5], dtype=np.float32).reshape(1, -1)  # Reshape to 1x6 matrix
Y = np.dot(set_w, X_train) + set_b
```

```python
Y
```

```python
y_out = [relu_activation(y) for y in Y.flatten()]  # Flatten Y and apply ReLU
```

```python
X_train = X_train.flatten()
```

```python
fig, ax = plt.subplots(1, 1)

# Scatter plot for data points
ax.scatter(X_train,y_out, marker='x', c='r', label="Data Points")

# Line connecting all points
ax.plot(X_train, y_out, linestyle='-', color='b', label="Connecting Line")

# Add legend, labels, and formatting
ax.legend(fontsize='xx-large')
ax.set_ylabel('Price (in 1000s of dollars)', fontsize='xx-large')
ax.set_xlabel('Size (1000 sqft)', fontsize='xx-large')

# Display the plot
plt.show()
```
