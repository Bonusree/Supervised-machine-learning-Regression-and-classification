```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1]).reshape(-1,1) 
```

```python
# Define the plot_data function
def plot_data(X, y, ax):
    # Flatten y to avoid indexing errors
    y = y.flatten()

    pos = y == 1  # Positive class (y=1)
    neg = y == 0  # Negative class (y=0)
    
    # Plot the positive class points
    ax.scatter(X[pos, 0], X[pos, 1], marker='x', c='red', label="Class 1")
    
    # Plot the negative class points
    ax.scatter(X[neg, 0], X[neg, 1], marker='o', s=100, facecolors='none', edgecolors='blue', lw=3, label="Class 0")
    
    # Add legend
    ax.legend()
```

```python
# Create the plot
fig, ax = plt.subplots(1, 1, figsize=(4, 4))

# Plot the data
plot_data(X, y, ax)

# Set axis limits and labels
ax.axis([0, 4, 0, 3.5])
ax.set_ylabel(r'$x_1$')
ax.set_xlabel(r'$x_0$')

# Show the plot
plt.show()
```

```python
# Define the sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Function to draw a vertical threshold line
def draw_vthresh(ax, thresh):
    ax.axvline(x=thresh, color="red", linestyle="--", label=f"Threshold z={thresh}")
    ax.legend()

# Plot sigmoid(z) over a range of values from -10 to 10
z = np.arange(-10, 11)

# Create the figure and axis
fig, ax = plt.subplots(1, 1, figsize=(5, 3))

# Plot z vs sigmoid(z)
ax.plot(z, sigmoid(z), c="b")

# Plot the x_train points as red 'x' markers on the sigmoid curve
ax.scatter(z, sigmoid(z), color='red', label='Data Points (x_train)')

# Set the title and labels
ax.set_title("Sigmoid function with x_train points")
ax.set_ylabel('sigmoid(z)')
ax.set_xlabel('z')

# Add the vertical threshold line at z = 0
draw_vthresh(ax, 0)

# Display the plot
plt.legend()
plt.show()
```

```python
# Define the plot_data function
def plot_data(X, y, ax):
    # Flatten y to avoid indexing errors
    y = y.flatten()

    # Separate positive and negative points for plotting
    pos = y == 1
    neg = y == 0

    # Plot the positive points (class 1)
    ax.scatter(X[pos, 0], X[pos, 1], marker='x', c='red', label="Class 1")

    # Plot the negative points (class 0)
    ax.scatter(X[neg, 0], X[neg, 1], marker='o', s=100, facecolors='none', edgecolors='blue', lw=3, label="Class 0")

    # Add legend
    ax.legend()
```

```python
# Generate the decision boundary
x0 = np.arange(0, 6)  # Values for x0
x1 = 3 - x0  # Decision boundary equation

# Create the plot
fig, ax = plt.subplots(1, 1, figsize=(5, 4))

# Plot the decision boundary
ax.plot(x0, x1, c="b", label="Decision Boundary")

# Fill the region below the line
ax.fill_between(x0, x1, alpha=0.2)

# Plot the original data points
plot_data(X, y, ax)

# Set axis labels
ax.set_ylabel(r'$x_1$')
ax.set_xlabel(r'$x_0$')

# Show the plot
plt.show()
```
