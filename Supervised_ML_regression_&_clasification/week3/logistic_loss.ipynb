```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_loss(f_x, y):
    """Calculate logistic loss."""
    return -y * np.log(f_x) - (1 - y) * np.log(1 - f_x)

# Generate values for f_x (predicted probabilities)
f_x = np.linspace(0.01, 0.99, 100)

# Calculate loss for y = 1 and y = 0
loss_y1 = logistic_loss(f_x, 1)
loss_y0 = logistic_loss(f_x, 0)

# Plotting
plt.plot(f_x, loss_y1, label='y = 1')
plt.plot(f_x, loss_y0, label='y = 0')
plt.title('Logistic Loss Function')
plt.xlabel('Predicted Probability (f(x))')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Define a bowl-shaped function (e.g., a simple quadratic function)
z = x**2 + y**2

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Set labels
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis (Cost)')

plt.title('3D Surface Plot of a Cost Function')
plt.show()
```

```python
# Define the training data
x_train = np.array([0., 1, 2, 3, 4, 5], dtype=np.longdouble)
y_train = np.array([0, 0, 0, 1, 1, 1], dtype=np.longdouble)

# Create a simple plot
plt.figure(figsize=(8, 5))
plt.scatter(x_train, y_train, color='blue', label='Data Points')
plt.plot(x_train, y_train, color='orange', linestyle='--', label='Decision Boundary')

# Set labels and title
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Simple Example Plot')
plt.xticks(np.arange(0, 6, 1))
plt.yticks([0, 1])
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

```python
# Create a grid of values for w (weight) and b (bias)
w = np.linspace(-5, 5, 100)
b = np.linspace(-5, 5, 100)
W, B = np.meshgrid(w, b)

# Define the logistic function
def logistic(z):
    return 1 / (1 + np.exp(-z))

# Calculate the logistic squared error for each combination of w and b
Z = np.zeros(W.shape)
for i in range(W.shape[0]):
    for j in range(W.shape[1]):
        predictions = logistic(W[i, j] * x_train + B[i, j])
        Z[i, j] = np.mean((predictions - y_train) ** 2)  # Mean squared error

# Create a 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(W, B, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('Weight (w)')
ax.set_ylabel('Bias (b)')
ax.set_zlabel('Logistic Squared Error')
plt.title('3D Plot of Logistic Squared Error')
plt.show()
```

```python
# Define the logistic function
def logistic(z):
    return 1 / (1 + np.exp(-z))

# Define the logistic cost function
def logistic_cost(w, b, x, y):
    predictions = logistic(w * x + b)
    cost = -np.mean(y * np.log(predictions) + (1 - y) * np.log(1 - predictions))
    return cost

# Create a grid of values for w (weight) and b (bias)
w = np.linspace(-5, 5, 100)
b = np.linspace(-5, 5, 100)
W, B = np.meshgrid(w, b)

# Calculate the logistic cost for each combination of w and b
C = np.zeros(W.shape)
for i in range(W.shape[0]):
    for j in range(W.shape[1]):
        C[i, j] = logistic_cost(W[i, j], B[i, j], x_train, y_train)

# Create a 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(W, B, C, cmap='viridis')

# Set labels and title
ax.set_xlabel('Weight (w)')
ax.set_ylabel('Bias (b)')
ax.set_zlabel('Logistic Cost')
plt.title('3D Plot of Logistic Cost Function')
plt.show()
```

```python


log_C = np.log(C + 1e-15)  # Adding a small value to avoid log(0)

# Create a 3D plot for log(logistic cost)
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(W, B, log_C, cmap='viridis')

# Set labels and title
ax.set_xlabel('Weight (w)')
ax.set_ylabel('Bias (b)')
ax.set_zlabel('Log(Logistic Cost)')
plt.title('3D Plot of Log(Logistic Cost) vs Weight and Bias')
plt.show()
```
